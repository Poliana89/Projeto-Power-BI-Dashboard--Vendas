# **Dashboard-Vendas**

**Objetivo**

Criar um dashboard onde irá mostrar os KPIs de vendas, custos e médias.

**Carregamento e Transformação**

Arquivos .csv clientes, pedidos, produtos, vendas, aplicado a limpeza de duplicidade e reconhecimento de cabeçalho nos arquivos, dentro do Power Query.

**Relacionamento das Tabelas**

Aplicado o relacionamento entre as tabelas com a cardinalidade nos relacionamentos, conforme regra de negócio.

# Análise e Construção do Relatório  Dashboard

Qual foi o total de valor venda considerando cada modo de envio dos pedidos?

*Foi utilizado o gráfico de cascata, trazendo as categorias de modo de envio e a soma total para cada categoria.*

Quais mercados tiveram o maior custo médio de envio dos produtos vendidos?

*Foi utilizado o gráfico Treemap, mostrando a média de custo de envio por categoria de mercado.*

A empresa tem como objetivo (meta) manter uma média de 350 para o valor de venda todos os meses, com o valor médio de venda. A empresa ficou abaixo ou acima da meta no mês de abril/2014?

*Utilizado o gráfico indicador (velocímetro), por média do total de vendas e adicionado a meta de 350, bastando filtrar por data do pedido para descobrir se a empresa bateu a meta de vendas.*

Considere que o lucro é equivalente a: valor venda - custo envio. Qual categoria de produto apresentou maior lucro médio ?

*Primeiro foi adicionado uma coluna calculada chamada Lucro, calculo:*

Lucro = Vendas[Valor Venda]-Vendas[Custo Envio]
Com  a coluna calculada (lucro) pronta , foi adicionado o gráfico de rosca, trazendo a média de lucro por categoria de produtos.

Qual foi o comportamento da margem de lucro ao longo do tempo?

Adicionamos a coluna MargemLucro na tabela venda e feito o cálculo, utilizando função Dax , com a seguinte formula :
MargemLucro = ROUND(DIVIDE(Vendas[Lucro],Vendas[Valor Venda]) *100,2)
Com a coluna calculada (MargemLucro) pronta, foi adicionado o gráfico de linha de tempo mostrando a  margem do lucro ao longo do tempo
